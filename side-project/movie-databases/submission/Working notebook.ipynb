### Creating the RLS table

```python
rls = title_episode["show_id"].tolist()
len(rls)
```

```python
import random
rls_dict = {}

for i in rls:
    rls_dict[i] = [random.randint(1,801) for i in range(20)]
    
rls_dict
```

```python
actors_id = []

for i in rls_dict.values():
    for x in i:
        actors_id.append(x)
```

```python
rls_id = []

for i in rls:
    for x in range(20):
        rls_id.append(i)
```

```python
print(len(actors_id))
print(len(rls_id))
```

```python
rls_table = pd.DataFrame({"shows_id": rls_id, "actors_id": actors_id})
rls_table.head()
```
