# SQL Database


```python
import pandas as pd 
import sqlite3 as sql
import pandas.io.sql as pds

dealers_df = pd.read_csv("src/Auction_Dealer.csv")
cars_df = pd.read_csv("src/Auction_Cars.csv")

#Check
print(len(dealers_df)) #num of data points same as dataset. good to go!
print(len(cars_df))
print(sum(dealers_df.isnull().sum()))#check missing values
print(sum(cars_df.isnull().sum()))#check missing values

#edit the col names
dealers_df.columns = dealers_df.columns.str.replace(".", "").str.replace(" ", "_")
cars_df.columns = cars_df.columns.str.replace(" ", "_").str.replace(".", "")


#Change Dtype
dealers_df["Registration_Date"] = pd.to_datetime(dealers_df["Registration_Date"])
dealers_df["Bidding_Date"] = pd.to_datetime(dealers_df["Bidding_Date"])
cars_df["Bidded_Date"] = pd.to_datetime(cars_df["Bidded_Date"])
```

```python
#Database Setup
# make sure of updating the path to a file
conn = sql.connect('auction.db')

# index = False to ensure the DataFrame row index is not written into the SQL tables
dealers_df.to_sql('dealers', con = conn, index = False) 
cars_df.to_sql('cars', con = conn, index = False)

db = 'auction.db'

def run_query(q):
    # Connect to the database
    with sql.connect(db) as conn:
        # Wrapper function to read_sql_query(): returns a DataFrame corresponding to the result set of 
        # the query string
        return pd.read_sql(q,conn)
 
#Function to show all the tables present in the database.

def show_tables():
    q = '''
        SELECT
            name
        FROM sqlite_master
        WHERE type IN ("table","view");
        '''
    return run_query(q)

# Function to count how many rows in each table
def get_table_row_count(tablename):
    q = '''
        SELECT
            COUNT(1)
        FROM %s;
        ''' % tablename
    return run_query(q)["COUNT(1)"][0]

tables = show_tables()
tables["row_count"] = [get_table_row_count(t) for t in tables["name"]]

tables
```

```python
run_query("""SELECT strftime('%Y-%m', Registration_Date) AS 'Year-Month', COUNT(Dealer_ID) AS 'No. of Dealers Registered' 
                 FROM dealers
                 GROUP BY 1""")
```

```python
run_query("""SELECT Dealer_ID AS 'Dealer ID', SUM(No_of_Cars_Won) AS 'No. of Cars Won'
              FROM dealers
              GROUP BY 1
              ORDER BY 2
              LIMIT 2; """)
```

```python
run_query("""SELECT Car_Brand AS 'Car Brand', Car_model AS 'Car Model', ROUND(AVG(Reserve_Price)) AS 'Average Reserve Price'
             FROM cars
             GROUP BY 1, 2
             HAVING 'Average Reserve Price' > 300000
          """)
```
