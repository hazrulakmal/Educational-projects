# Binary Classification 

Logistic Regression is used as a model to predict wheather a telco customer will churn. Below is the code used to analyse the data and to make customer classification.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

# **Data Cleaning**

```python
df = pd.read_csv("Telco-Customer-Churn.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df[df["TotalCharges"].isnull()]
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")

string_col = list(df.dtypes[df.dtypes == "object"].index)
for col in string_col:
    df[col] = df[col].str.lower().str.replace(" ", "_")
    
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.head()
```

```python
df.churn = (df.churn == "yes").astype(int)
```

```python
df.head()
```

## Setting up a validation framework

```python
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)
len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop= True)
df_val = df_val.reset_index(drop= True)
df_test = df_test.reset_index(drop= True)

y_train = df_train.churn.values
y_val = df_val.churn.values
y_test = df_test.churn.values

del df_train["churn"]
del df_val["churn"]
del df_test["churn"]
```

```python
df_train.head()
```

## EDA

```python
df_full_train = df_full_train.reset_index(drop=True)
df_full_train.isnull().sum()
```

```python
global_churn = df_full_train["churn"].value_counts(normalize=True).iloc[1]
global_churn
```

Churn rate is 26%

```python
numeric = ["tenure", "monthlycharges", 'totalcharges']
categorical = ['gender', 'seniorcitizen', 'partner', 'dependents', 'phoneservice', 'multiplelines', 'internetservice',
       'onlinesecurity', 'onlinebackup', 'deviceprotection', 'techsupport',
       'streamingtv', 'streamingmovies', 'contract', 'paperlessbilling',
       'paymentmethod']
```

## Feature Importance

### Churn Rate

```python
from IPython.display import display

for col in categorical:
    print(col)
    df_group = df_full_train.groupby(col)["churn"].agg(["mean", "count"])
    df_group["diff"] = global_churn - df_group["mean"] 
    df_group["risk"] = df_group["mean"] / global_churn
    display(df_group)
```

### Mutual Information 
concept from information theory, it tells us how much we can learn about one variable if we know the value from another

```python
from sklearn.metrics import mutual_info_score

def mutual_info_scores(series):
    return mutual_info_score(series, df_full_train.churn)

mutual_info = df_full_train[categorical].apply(mutual_info_scores)
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).index
```

### Correlation

```python
df_full_train[numeric].corrwith(df_full_train.churn).abs().sort_values(ascending=False)
```

```python
from sklearn.feature_extraction import DictVectorizer

selected_categorical = ['contract', 'onlinesecurity', 'techsupport', 'internetservice', 'onlinebackup', 'paymentmethod', 'deviceprotection', 'streamingtv','streamingmovies']
selected_numeric = ["tenure", "totalcharges",]

def transform_cols(dataframe):
    dv = DictVectorizer(sparse=False)
    dic = dataframe[selected_categorical+selected_numeric].to_dict(orient="records")
    dv.fit(dic)
    return dv.transform(dic)

X_train = transform_cols(df_train)
X_val = transform_cols(df_val)
X_test = transform_cols(df_test)
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(X_train, y_train)

val_preds = model.predict(X_val)
(val_preds == y_val).mean()
```

```python
val_preds_2 = model.predict_proba(X_val)[:, 1]
churn_decisions = val_preds_2 >= 0.5
(churn_decisions.astype(int) == y_val).mean()
```

```python
train_preds = model.predict(X_train)
(train_preds == y_train).mean()
```

## Use the model

```python
X_full = transform_cols(df_full_train)
y_full = df_full_train.churn.values
```

```python
lr = LogisticRegression(solver='liblinear', random_state=1)
lr.fit(X_full, y_full)

test_preds = lr.predict(X_test)
(test_preds == y_test).mean()
```

Accurancy of the model tested on the test dataset is 81%
